# Merge

```python
di_ct = pd.DataFrame({"A":[1,2,3,5],"B":[11,22,33,44]})
```

```python
di_ct1 = pd.DataFrame({"A":[1,2,3,6],"C":[55,66,77,88]})
pd.merge(di_ct,di_ct1, on = "A")
```

```python
pd.merge(di_ct,di_ct1, on  = "A")
```

```python
pd.merge(di_ct,di_ct1, how = "inner")
```

```python
pd.merge(di_ct,di_ct1, how = "left")
```

```python
pd.merge(di_ct,di_ct1, how = "right")
```

```python
pd.merge(di_ct,di_ct1, how = "outer",indicator = True)
```

```python
pd.merge(di_ct,di_ct1,left_index = True,right_index= True, suffixes = ("Zami","hi"))
```

```python
di_ct3 = pd.DataFrame({"A":[1,2,3,5],"B":[11,22,33,44]})
di_ct2 = pd.DataFrame({"A":[1,2,3,9],"B":[11,45,39,40]})
```

```python
pd.merge(di_ct3,di_ct2,left_index = True,right_index= True, suffixes = ("Zami","hi")) #suffixes to give the names of the columns like below zami for A and Hi For B
```
